## Pregunta TS1.8 ¿Cómo se podría obtener la misma salida pero utilizando una sola transformación y sin realizar la unión?

Se puede obtener la misma salida juntando ambas condiciones en la función lambda mediante el operador "or".

```python
log = sc.parallelize(['E: e21', 'I: i11', 'W: w12', 'I: i11', 'W: w13', 'E: e45'])

infos = log.filter(lambda elemento: elemento[0]=='I')
errors = log.filter(lambda elemento: elemento[0]=='E')

inferr = infos.union(errors)

print (inferr.collect())
```

```python
#¿Cómo se podría obtener la misma salida pero utilizando una sola transformación y sin realizar la unión?
rdd_log = sc.parallelize(['E: e21', 'I: i11', 'W: w12', 'I: i11', 'W: w13', 'E: e45'])

print(rdd_log.filter(lambda x: x[0]=="I" or x[0]=="E").collect())
```

Al cambiar el orden de la concatenación de palabras mediante un guión, se unen dichas palabras en orden inverso, es decir, se pasa de "hola-que-tal-bien" a "bien-tal-que-hola".

```python
r = sc.parallelize([('A', 1),('C', 4),('A', 1),('B', 1),('B', 4)])
rr = r.reduceByKey(lambda v1,v2:v1+v2)
print (rr.collect())
```

Mediante el método reduceByKey se agregan los pares clave valor, agregrandose los valores que comparten clave mediante la función especificada, en este caso, mediante la suma de dichos valores

```python
r = sc.parallelize([('A', 1),('C', 4),('A', 1),('B', 1),('B', 4)])
rr1 = r.reduceByKey(lambda v1,v2:v1+v2)
print (rr1.collect())
rr2 = rr1.reduceByKey(lambda v1,v2:v1)
print (rr2.collect())
```

Las primeras 3 líneas de la celda son idénticas a las de la anterior, la diferencia estña en las últimas 2 líneas, en las que el rdd resultado de las operaciones anteriores es pasado como entrada un nuevo reduceByKey en el que la función especificada busca quedarse con el primer valor del par, cosa que en este caso, resulta dar exactamente el mismo output al rdd inicial, luego probablemente no se haya conseguido lo esperado con esta operación.

Para realizar la alternativa al reduceByKey, se debe en primer lugar transformar el tipo de los elemento agrupados de forma que pasen de ser un iterador a una lista, y posteriormente gracias al método map y la función que suma los elementos de cada clave "sum" se obtiene el resultado deseado.

```python
rdd1 = sc.parallelize([('A',1),('B',2),('C',3)])
rdd2 = sc.parallelize([('A',4),('B',5),('C',6)])

rddjoin = rdd1.join(rdd2)

print (rddjoin.collect())
# Prueba a cambiar las claves del rdd1 y rdd2 para ver cuántos elementos se crean
```

```python
rdd1 = sc.parallelize([('A',1),('B',2),('C',3)])
rdd2 = sc.parallelize([('A',4),('A',5),('B',6),('D',7)])

rddjoin = rdd1.join(rdd2)
print(rddjoin.collect())

# Modifica join por leftOuterJoin, rightOuterJoin y fullOuterJoin ¿Qué sucede? 
rdd_leftjoin = rdd1.leftOuterJoin(rdd2)
print(rdd_leftjoin.collect())

rdd_rightjoin = rdd1.rightOuterJoin(rdd2)
print(rdd_rightjoin.collect())

rdd_fulljoin = rdd1.fullOuterJoin(rdd2)
print(rdd_fulljoin.collect())
```

Con el join solo se crean aquellos pares que comparten clave tanto en el rdd1 como en el rdd2.

Con el leftOuterJoin además de los pares que comparten clave, también se crean los pares del rdd de la izquiera, en este caso, del rdd1 => ('C', (3, None)).

Con el RightOuterJoin además de los pares que comparten clave, también se crean los pares del rdd de la derecha, en este caso, del rdd2 => ('D', (None, 7)).

Y con el fullOuterJoin además de los pares que comparten clave, también se crean los pares de ambos rdd (tanto de la izquierda como de la derecha) => ('C', (3, None)) y ('D', (None, 7)).

```python
rdd = sc.parallelize([('A',1),('B',2),('C',3),('A',4),('A',5),('B',6)])

res = rdd.sortByKey(False)

print (res.collect())
```

# Procesando el QUIJOTE

```python
quijote = sc.textFile("/content/gdrive/My Drive/PGDE/quijote.txt")
quijote.take(10)
```

Transformaciones

```python
charsPerLine = quijote.map(lambda s: len(s))
print(charsPerLine.take(10))
```

Mediante el uso de map y la función len() especificada, nos permite saber la longitud de cada una de las líneas del quijote. Por tanto, la cantidad de elementos inciales coincide con la del RDD final (N=M).

```python
allWords = quijote.flatMap(lambda s: s.split())
print(allWords.take(10))
```

Mediante el uso de flatmap y la función split() especificada nos permite separar cada linea del quijote en sus distintas palabras, es decir, tokenizamos el texto. Por tanto, la cantidad de elementos iniciales es menor a la del RDD output (N<M).

```python
allWordsNoArticles = allWords.filter(lambda a: a.lower() not in ["el", "la"])
print(allWordsNoArticles.take(10))
```

En base al RDD allWords y gracias al método filter con la función que elimina los articulos "el" y "la" sin tener en cuanta mayúsculas o minúsculas (case insensitive) nos permite obtener un RDD con el quijote tokenizao pero sin los mencionados articulos. Por tanto, el RDD incial es mayor al final (N>M).

```python
allWordsUnique = allWords.map(lambda s: s.lower()).distinct()
print(allWordsUnique.take(10))
```

En base al RDD allWords y gracias al método map y la función especificada lower() se transforma todos los tokens a minúsculas para posteriormente mediante el método distinct se descarten aquellos tokens repetidos. Por tanto, el RDD incial es mayor al final (N>M).

```python
sampleWords = allWords.sample(withReplacement=True, fraction=0.2, seed=666)
print(sampleWords.take(10))
```

En base al RDD allWords, se da una muestra de dichos tokens en la que se permite elementos repetidos, en la que los elementos son seleccionados o no con una probabilidad del 20% y con un seed que nos permite repetir la misma muestra en las distintas ejecuciones. Por tanto, N>M.

```python
weirdSampling = sampleWords.union(allWordsNoArticles.sample(False, fraction=0.3))
print(weirdSampling.take(10))
```

Se hace una unión entre la muestra sampleWors y otra muestra en la que no se permiten repetidos, en la que los elementos son seleccionados o no con un 30% de probabilidad, pero esta vez sin seed. Por tanto N<M.

Similar a la operación anteriormente realizada con la diferencia de que está vez se hace una petición web para obtener la parte 2 del quijote y que se añade la función decode como parte de los preprocesos para asegurarnos que la codificación es utf-8.

```python
allWords.take(10)
```

```python
allWords2.take(10)
```

```python
words = allWords.map(lambda e: (e,1))
words2 = allWords2.map(lambda e: (e,1))

words.take(10)
```

Se hace una asociación clave-valor, de las distintas palabras de ambas partes del quijote, en la que cada palabra (repetida o no) se le asigna el valor 1. 

```python
frequencies = words.reduceByKey(lambda a,b: a+b)
frequencies2 = words2.reduceByKey(lambda a,b: a+b)
frequencies.takeOrdered(10, key=lambda a: -a[1])
```

Tras la asignación clave-valor anteriormente realizada, mediante el método reduceByKey, se juntan los distintos tokens repetidos, de forma que se cuenta la frecuencia de aparición de cada uno de los elementos de los textos.

```python
res = words.groupByKey().takeOrdered(10, key=lambda a: -len(a))
res # To see the content, res[i][1].data
```

Se agrupan las palabras de la primera parte del quijote por su clave, es decir, por la palabra en sí, y su valor pasa a ser un iterador de Spark que contiene tantos elementos como veces aparezca en el texto dicha palabra.

```python
joinFreq = frequencies.join(frequencies2)
joinFreq.take(10)
```

Mediante el método join se permite comparar la frequencia de aparición de las distintas palabras de cada una de las partes del quijote.

```python
joinFreq.map(lambda e: (e[0], (e[1][0] - e[1][1])/(e[1][0] + e[1][1]))).takeOrdered(10, lambda v: -v[1])
```

Se saca la proporción de la frecuencia de aparición de cada una de las palabras de la primera parte del quijote con respecto a la segunda, dividido entre la frecuencia de aparición total de dicha palabra. Posteriormente se imprime en pantalla los 10 primeros elementos de mayor a menor proporción.

Esta es una forma para de un vistazo ver qué palabras se usan más en cada parte del quijote.

```python
joinFreq.map(lambda e: (e[0], (e[1][0] - e[1][1])/(e[1][0] + e[1][1]))).takeOrdered(10, lambda v: +v[1])
```

Se realizo lo mismo que en la última celda, solo que esta vez ordenando de menor a mayor.

```python
import time
```

```python
start = time.time()
joinFreq.map(lambda e: (e[0], (e[1][0] - e[1][1])/(e[1][0] + e[1][1]))).takeOrdered(10, lambda v: -v[1]), joinFreq.map(lambda e: (e[0], (e[1][0] - e[1][1])/(e[1][0] + e[1][1]))).takeOrdered(10, lambda v: +v[1])
print(time.time() - start)
```

```python
start = time.time()
result = joinFreq.map(lambda e: (e[0], (e[1][0] - e[1][1])/(e[1][0] + e[1][1])))
result.cache()
result.takeOrdered(10, lambda v: -v[1]), result.takeOrdered(10, lambda v: +v[1])
print(time.time() - start)
```

Como en la segunda celda solo se realiza la operación de mapeo una vez y luego se guarda en caché para posteriormente ordenar ascendentemente y descendentemente, nos ahorramos esta segunda operación de mapeo, luego en la teoría esto hace que la operación sea más eficiente. Además, se ha comprobado de forma práctica viendo que la segunda celda tarda en realizar la misma operación significativamente menos tiempo.
